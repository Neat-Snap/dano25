```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None) 
```

```python
df = pd.read_excel("df.xlsx")
df
```

```python
df["review_day"] = df["review_dttm"].dt.date
df["review_time"] = df["review_dttm"].dt.time
df["review_month"] = df["review_dttm"].dt.month

df["finish_day"] = df["finish_dttm"].dt.date
df["finish_time"] = df["finish_dttm"].dt.time
df["finish_month"] = df["finish_dttm"].dt.month
```

```python
null_fillable = ["children_cnt", "subscription_important_flg", "new_flg", "influencer_flg"]
df[null_fillable] = df[null_fillable].fillna(0)

df["gender_cd"] = df["gender_cd"].fillna("M")
```

```python
education_mapping = {
    "GRD": 0, # Graduate
    "SCH": 1, # School
    "UGR": 2, # Undergraduate
    "PGR": 3, # Postgraduate
    "ACD": 4, # Academic
    "MGR": 5  # Managerial
}

df["education_level_cd"] = df["education_level_cd"].map(education_mapping)

df.marital_status_cd.value_counts()
marital_mapping = {
    "UNM": 0, # Unmarried
    "MAR": 1, # Married
    "DIV": 2, # Divorced
    "WID": 3, # Widowed
    "CIV": 4  # Civil Union
}   

df["marital_status_cd"] = df["marital_status_cd"].map(marital_mapping)

df["citizenship_country"] = df["citizenship_country"].fillna("РФ")

df["review_mark"] = df["review_mark"].replace("без оценки (негативный)", 1)
df["review_mark"] = df["review_mark"].replace("без оценки (нейтральный)", 3)
df["review_mark"] = df["review_mark"].replace("без оценки (позитивный)", 5)
df["review_mark"] = df["review_mark"].astype(int)

df["review_emotion"] = df["review_emotion"].replace("Позитивный", 1)
df["review_emotion"] = df["review_emotion"].replace("Негативный", 0)
df.review_emotion.value_counts()

for i in df.drop(["finish_dttm", "review_dttm", "id_client", "review_text"], axis=1).columns:
    print(f"{i}: {df[i].nunique()} unique values")
    print(df[i].value_counts())
    print("----------")
```
