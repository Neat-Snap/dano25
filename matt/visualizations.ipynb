```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)


import seaborn as sns
import matplotlib.pyplot as plt

# 1. Цветовая палитра (взята со смайликов на слайде)
colors = {
    'bg': '#EFE8E8',        # Фон слайда (светло-бежевый)
    'text': '#513838',      # Основной текст (темно-коричневый)
    
    # Акцентные цвета смайловb
    'red': '#D64045',       # Злой (красный)
    'blue': '#1064A8',      # Грустный (синий)
    'cyan': '#009FB7',      # Милый (бирюзовый)
    'green': '#258E62',     # Скептический (зеленый)
    'purple': '#582CA1',    # Фиолетовый
    'orange': '#F57F25',    # Удивленный (оранжевый)
    'yellow': '#FCC427',    # Веселый (желтый)
    'pink': '#C93679',      # Розовый
}

# Создаем список цветов для автоматического циклирования в графиках
custom_palette = [
    colors['blue'], colors['red'], colors['green'], 
    colors['orange'], colors['purple'], colors['cyan']
]

# 2. Настройки параметров (RC params)
custom_params = {
    # Фон
    "figure.facecolor": colors['bg'],
    "axes.facecolor": colors['bg'],
    "savefig.facecolor": colors['bg'],
    
    # Текст и шрифты
    "text.color": colors['text'],
    "axes.labelcolor": colors['text'],
    "xtick.color": colors['text'],
    "ytick.color": colors['text'],
    "font.family": "sans-serif",  # Можно заменить на конкретный шрифт презентации, например 'Arial' или 'Roboto'
    
    # Сетка (бледная и тонкая)
    "axes.grid": True,
    "grid.color": colors['text'],
    "grid.alpha": 0.15,           # Очень прозрачная
    "grid.linestyle": "-",
    "grid.linewidth": 0.8,
    
    # Границы (Spines) - убираем лишнее, оставляем тонкие
    "axes.spines.left": True,
    "axes.spines.bottom": True,
    "axes.spines.right": False,   # Убираем правую рамку
    "axes.spines.top": False,     # Убираем верхнюю рамку
    "axes.edgecolor": colors['text'],
    "axes.linewidth": 0.8,
    
    # Линии графиков
    "lines.linewidth": 1.5,       # Тонкие линии, как просили
    "lines.markersize": 6,
}

# 3. Применение стиля
sns.set_theme(
    context="talk",           # Масштаб (paper, notebook, talk, poster) - talk хорош для слайдов
    style="whitegrid",        # Базовый стиль
    rc=custom_params,         # Наши кастомные настройки
    palette=custom_palette    # Наша палитра
)
```

```python
df = pd.read_csv("cleaned.csv")
df
```

```python
# Check dataset shape and column info
print(f"Dataset shape: {df.shape}")
print(f"\nColumn names and types:")
df.info()
```

```python
# Create a comprehensive dataset overview visualization
fig, ax = plt.subplots(figsize=(14, 8))

# Prepare data for visualization
total_records = len(df)
total_columns = len(df.columns)
date_range = f"{df['review_dttm'].min()} - {df['review_dttm'].max()}"
completeness = (df.notna().sum() / len(df) * 100).mean()

# Key metrics
key_metrics = {
    'Total Reviews': f"{total_records:,}",
    'Time Period': date_range,
    'Unique Clients': f"{df['id_client'].nunique():,}",
    'Products': f"{df['product'].nunique()}",
    'Business Lines': f"{df['business_line'].nunique()}",
    'Review Sources': f"{df['review_source'].nunique()}",
    'Avg Completeness': f"{completeness:.1f}%",
    'Data Columns': f"{total_columns}"
}

# Create text-based info panel
ax.axis('off')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)

# Title
ax.text(5, 9.2, 'Dataset Overview', 
        ha='center', va='top', fontsize=32, fontweight='bold', color=colors['text'])

# Subtitle
ax.text(5, 8.5, f'Bank Customer Reviews Analysis', 
        ha='center', va='top', fontsize=18, color=colors['text'], alpha=0.8)

# Create boxes for metrics
y_start = 7.2
box_height = 0.9
box_width = 4.5
spacing = 0.3

metrics_list = list(key_metrics.items())
color_list = [colors['blue'], colors['red'], colors['green'], colors['orange'], 
              colors['purple'], colors['cyan'], colors['yellow'], colors['pink']]

for i, ((label, value), color) in enumerate(zip(metrics_list, color_list)):
    row = i // 2
    col = i % 2
    
    x = 0.5 + col * (box_width + spacing)
    y = y_start - row * (box_height + spacing)
    
    # Draw colored rectangle
    rect = plt.Rectangle((x, y - box_height), box_width, box_height, 
                         facecolor=color, alpha=0.2, edgecolor=color, linewidth=2)
    ax.add_patch(rect)
    
    # Add label
    ax.text(x + box_width/2, y - 0.25, label, 
           ha='center', va='center', fontsize=14, color=colors['text'], fontweight='600')
    
    # Add value
    ax.text(x + box_width/2, y - 0.65, value, 
           ha='center', va='center', fontsize=18, color=color, fontweight='bold')

# Add a small note at the bottom
ax.text(5, 0.5, 'Comprehensive customer feedback database covering multiple products and channels', 
        ha='center', va='center', fontsize=12, color=colors['text'], alpha=0.7, style='italic')

plt.tight_layout()
plt.savefig('dataset_overview.png', dpi=300, bbox_inches='tight', facecolor=colors['bg'])
plt.show()

print("✓ Visualization saved as 'dataset_overview.png'")
```
