```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dfc.csv")
df = df[df["review_source"] == 'нет']
```

```python
df["review_theme"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
counts = df["review_theme"].value_counts().reset_index()
counts.columns = ['theme', 'count']
plt.figure(figsize=(12, 6))
sns.set_theme(style="whitegrid")
ax = sns.barplot(x='count', y='theme', data=counts, palette='viridis')
for i, v in enumerate(counts['count']):
    ax.text(v + 5, i, str(v), color='black', va='center', fontweight='bold')
plt.title('Распределение отзывов по темам', fontsize=14)
plt.xlabel('Количество', fontsize=12)
plt.ylabel('Тема', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('dfc.csv')
target_theme = 'тарифы и условия'
subset = df[df['review_theme'] == target_theme]
subset_clean = subset[subset['reason'] != 'Не определено']
reason_counts = subset_clean['reason'].value_counts().head(15)

plt.figure(figsize=(10, 6))
sns.barplot(x=reason_counts.values, y=reason_counts.index, palette='viridis')

plt.title(f'Топ-15 причин обращений по теме "{target_theme}"\n(исключая "Не определено")')
plt.xlabel('Количество обращений')
plt.ylabel('Причина')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
