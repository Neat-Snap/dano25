```python
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', None) 
```

```python
df = pd.read_excel('df.xlsx')
df
```

```python
df = df[df['company'] == 'Мобайл']
```

```python
df['no_mark_flag'] = df['review_mark'].str.contains('без оценки', case=False, na=False)

df["review_mark"].replace({'без оценки (негативный)': '2', 'без оценки (позитивный)': '4', 'без оценки (нейтральный)': '3'}, inplace=True)
df['review_mark_numeric'] = pd.to_numeric(df['review_mark'], errors='coerce')

display(df['review_mark_numeric'].value_counts())
display(df['no_mark_flag'].value_counts())
```

```python
emotion_map = {
    "Негативный": 0,
    "Позитивный": 1,
}

df["review_emotion"] = df["review_emotion"].map(emotion_map)

product_group_map = {
    'мобайл': "Мобильная связь",
    'sim': 'Мобильная связь',
    'Не определено': 'Не определено',
    'не определено': 'Не определено',
}
df["product"] = df["product"].map(product_group_map)
```

```python
df["product"].value_counts()
```

```python
df.to_csv("cleaned.csv", index=False)
```
